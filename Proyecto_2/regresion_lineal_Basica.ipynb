```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#Data set de datos ficticios de precios
np.random.seed(42)
superficie = np.random.randint(50,150,100)#Metros cuadrados
precio = superficie*3000 + np.random.normal(0,10000, 100) #Precios con ruido

df = pd.DataFrame({'superficie': superficie, 'precio': precio})
print(df.head())
```

```python
#Ajustar modelo de regresion
from sklearn.linear_model import LinearRegression

data_x = df[['superficie']]
data_y = df['precio']

modelo = LinearRegression()
modelo.fit(data_x, data_y)

print(f'Coeficiente:{modelo.coef_[0]}, Intercepto:{modelo.intercept_}')
```

```python
#Visualizacion de resultados
plt.scatter(data_x, data_y, color='blue')
plt.plot(data_x, modelo.predict(data_x), color='red')
plt.show()
```
