```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Carga y preparacion de datos:
data = pd.read_csv('ventas.csv', parse_dates=['Date'], index_col='Date')
print(data.head())
```

```python
data.plot()
plt.show()
```

```python
#Calcular promedio movil (Moving Average)
data['MA_3'] = data['Sales'].rolling(window=3).mean()
data[['Sales', 'MA_3']].plot()
plt.show()
```

```python
#Division del dataSet para prediccion
train = data[:80]
test = data[80:]
plt.plot(train, label = 'Entrenamiento')
plt.plot(test, label='Prueba')
plt.legend()
plt.show()
```
