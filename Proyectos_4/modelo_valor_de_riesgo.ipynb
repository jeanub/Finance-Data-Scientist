```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
rendimientos = np.random.normal(0.001, 0.02, 1000)#Simulacion de rendimientos
df = pd.DataFrame(rendimientos, columns=['rendimientos'])
```

```python
var_95 = np.percentile(df['rendimientos'],5)
print(f'Var al 95%:{var_95:.4f}')
```

```python
#Grafica de los rendimientos para identificar riesgos
df['rendimientos'].plot(kind='hist',bins=50)
plt.axvline(var_95,color='red', linestyle='dashed', linewidth=2)
plt.show()
```
